```python
import random

def calcolatore_tempo():
    num_serie = int(input("Quanti show vuoi vedere? "))
    total = 0

    for i in range(1, num_serie + 1):
        num_ep = int(input(f"Serie {i} - Inserisci quanti episodi ha la serie: "))
        tempo_per_ep = int(input(f"Serie {i} - Inserisci (in minuti) quanto dura ogni episodio: "))

        total += num_ep * tempo_per_ep

    num_ore = total // 60

    print("------------------------------------------")
    print(f"Ti servono {total} minuti per completare tutto.")
    print("(")

    if total >= 24 * 60:
        if num_ore // 24 == 1:
            print(f"{num_ore // 24} giorno, ", end="")
        else:
            print(f"{num_ore // 24} giorni, ", end="")

    if num_ore != 0:
        if total >= 24 * 60:
            if num_ore % 24 == 1:
                print(f"{num_ore % 24} ora, ", end="")
            else:
                print(f"{num_ore % 24} ore, ", end="")

        if num_ore == 1 and num_ore < 24:
            print(f"{num_ore} ora, ", end="")
        elif num_ore != 1 and num_ore < 24:
            print(f"{num_ore} ore, ", end="")

    if total % 60 == 1:
        print(f"{total % 60} minuto.", end="")
    else:
        print(f"{total % 60} minuti.", end="")

    print(")")
    print("------------------------------------------")

    num_giorni = int(input("In quanti giorni vuoi finire tutto? "))
    print("------------------------------------------")
    print("Per vedere tutto equamente in diversi giorni ti ci vuole: ")

    for j in range(1, num_giorni + 1):
        num_minuti_giornalieri = (total % (60 * j)) // j
        num_ore_giornaliere = num_ore // j

        if num_ore_giornaliere == 0:
            num_minuti_giornalieri = total // j

        ore_plur_sing = "ora" if num_ore_giornaliere == 1 else "ore"
        min_plur_sing = "minuto" if num_minuti_giornalieri == 1 else "minuti"

        if num_ore_giornaliere != 0:
            print(f"{num_ore_giornaliere} {ore_plur_sing}", end=" ")

        if num_ore_giornaliere != 0 and num_minuti_giornalieri != 0:
            print("e ", end="")

        if num_minuti_giornalieri != 0:
            print(f"{num_minuti_giornalieri} {min_plur_sing}", end=" ")

        if j == 1:
            print("in 1 Giorno. ", end="")
        else:
            print(f"in {j} Giorni. ", end="")

        if total // j > 24 * 60:
            print("Impossibile! ", end="")
        elif total // j > 18 * 60:
            print("Tecnicamente e' possibile... se rinunci alla tua sanita' mentale.", end="")
        elif total // j > 12 * 60:
            print("NO. Non pensarci nemmeno. Neanche per scherzo.", end="")
        elif total // j > 8 * 60:
            print("Ok, fattibile, ma che ne dici di prendertela con piu' calma?", end="")
        elif total // j > 6 * 60:
            print("Mmh... Richiede parecchio impegno ma si puo' fare.", end="")
        elif total // j > 4 * 60:
            print("Decisamente possibile... cosa stai aspettando?!", end="")
        elif total // j <= 4 * 60:
            print("Pfft... e questa la chiami sfida?", end="")
        print()

    continua = input("Vuoi continuare? digita: (si/no) ")
    return continua

if __name__ == "__main__":
    print("Benvenuto al calcolatore di tempo necessario!")
    continua = "si"
    while continua.lower() == "si":
        continua = calcolatore_tempo()
```
